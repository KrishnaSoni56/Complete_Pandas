# Attrition Data Analysis

```python
import numpy as np
import pandas as pd

df = pd.read_csv("D:\\Feature_csv_Files\\Attrition.csv")

df
```

```python
df.shape
```

# Department Name of employee that layoffs

```python
# filter yes category of attrition column
attr_data = df[df["Attrition"] == "Yes"]

# No of employee layoff with respective department 
department_attrition = attr_data.groupby('Department').size().reset_index(name = "Employee_Count")
department_attrition
```

# Education Field of employee that layoff

```python
edu_attr = attr_data.groupby('EducationField').size().reset_index(name = "Employee_Count")
edu_attr
```

# Marital Status of Layoffs Employee

```python
Marital_Status_attr = attr_data.groupby('MaritalStatus').size().reset_index(name = "Employee_Count")
Marital_Status_attr
```

# Overtime Employee layoff

```python
Overtime_Employee_attr = attr_data.groupby('OverTime').size().reset_index(name = "Employee_Count")
Overtime_Employee_attr
```

# Job Satisfation of Attrition Employees

```python
Job_satisfation_attr = attr_data.groupby('JobSatisfaction').size().reset_index(name = "Employee_Count")
Job_satisfation_attr
```

# No of years spend by employee that layoff

```python
Emp_YearsAtCompany = attr_data.groupby('YearsAtCompany').size().reset_index(name= "Count")
Emp_YearsAtCompany
```

# Employee layoff due to Distance

```python
Emp_distance = attr_data.groupby("DistanceFromHome").size().reset_index(name= "Count")
Emp_distance
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.io as pio
import plotly.graph_objects as go
pio.templates.default = "plotly_dark"
```

```python
print(df.isnull().sum())
```

```python
sns.distplot(df['Age'] )
plt.title('Distribution of Age') 
plt.show() 
```

```python
# Calculate the count of attrition by department
attrition_by = attr_data.groupby(['Department']).size().reset_index(name='Count')

# Create a donut chart
fig = go.Figure(data=[go.Pie(
    labels=attrition_by['Department'],
    values=attrition_by['Count'],
    hole=0.4,
    marker=dict(colors=['#3CAEA3', '#F6D55C']),
    textposition='inside'
)])

# Update the layout
fig.update_layout(title='Attrition by Department', font=dict(size=16), legend=dict(
    orientation="h", yanchor="bottom", y=1.02, xanchor="right", x=1
))

# Show the chart
fig.show()
```

```python
attrition_by = attr_data.groupby(['EducationField']).size().reset_index(name='Count')

# Create a donut chart
fig = go.Figure(data=[go.Pie(
    labels=attrition_by['EducationField'],
    values=attrition_by['Count'],
    hole=0.4,
    marker=dict(colors=['#3CAEA3', '#F6D55C']),
    textposition='inside'
)])

# Update the layout
fig.update_layout(title='Attrition by Educational Field', font=dict(size=16), legend=dict(
    orientation="h", yanchor="bottom", y=1.02, xanchor="right", x=1
))

# Show the chart
fig.show()
```

```python
attrition_by = attr_data.groupby(['YearsSinceLastPromotion']).size().reset_index(name='Count')
attrition_by
```

```python
attrition_by = attr_data.groupby(['Gender']).size().reset_index(name='count')
attrition_by

# Create a donut chart
fig = go.Figure(data=[go.Pie(
    labels=attrition_by['Gender'],
    values=attrition_by['count'],
    hole=0.4,
    marker=dict(colors=['#3CAEA3', '#F6D55C']),
    textposition='inside'
)])

# Update the layout
fig.update_layout(title='Attrition by Gender', font=dict(size=16), legend=dict(
    orientation="h", yanchor="bottom", y=1.02, xanchor="right", x=1
))

# Show the chart
fig.show()
```
